## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
 
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
mouse_metadata.head()
# Combine the data into a single dataset
merge_df= pd.merge(mouse_metadata, study_results, on='Mouse ID')
# Display the data table for preview
merge_df.head(3)
```

```python
# Checking the number of mice.
merge_df['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated = merge_df[merge_df.duplicated(['Mouse ID','Timepoint'])]
duplicated
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# we threw out mouse g989
clean_df = merge_df[(merge_df['Mouse ID']!='g989')]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame. after removing the duplicate mouse id
clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, ...
#and SEM of the tumor volume for each regimen

drug_regimen_group = clean_df.groupby('Drug Regimen')
mean_drug_regimen=drug_regimen_group.mean()['Tumor Volume (mm3)']
median_drug_regimen = drug_regimen_group.median()['Tumor Volume (mm3)']
variance_drug_regimen = drug_regimen_group.var()['Tumor Volume (mm3)']
std_drug_regimen = drug_regimen_group.std()['Tumor Volume (mm3)']
sem_drug_regimen = drug_regimen_group.sem()['Tumor Volume (mm3)']

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drugregimen_summary_df =pd.DataFrame({'Mean':mean_drug_regimen,'Median':median_drug_regimen, 
                                      'Variance':variance_drug_regimen
                                     ,'Standard Deviation':std_drug_regimen,'SEM':sem_drug_regimen})
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drugregimen_summary_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
drug_regimen_group.agg(['mean','median','var','std','sem'])['Tumor Volume (mm3)']
```

## Quartiles, Outliers and Boxplots

```python
tumor_max = clean_df.groupby('Mouse ID').max().reset_index()
merged = tumor_max[['Mouse ID','Timepoint']].merge(clean_df,on=['Mouse ID','Timepoint'])
drugs = ['Capomulin','Ramicane','Infubinol','Ceftamin']

tumor_list = []

for drug in drugs:
    tumor_data = merged[merged["Drug Regimen"] == drug]['Tumor Volume (mm3)']
    iqr = tumor_data.quantile(.75) - tumor_data.quantile(.25)
    lower = tumor_data.quantile(.25) - (1.5*iqr)
    upper = tumor_data.quantile(.75) + (1.5*iqr)
    
    tumor_list.append(tumor_data)
    
    print(f'{drug} potential outliers: {tumor_data[(tumor_data<lower) | (tumor_data>upper)]}')
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_df=clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin',:]
s185plot=capomulin_df.loc[capomulin_df['Mouse ID'] == 's185',:][['Timepoint','Tumor Volume (mm3)']]
s185plot=s185plot.set_index('Timepoint')
plt.plot(s185plot.index, s185plot['Tumor Volume (mm3)'])
plt.title('Capomulin: Timepoint vs Tumor Volume for MouseID: s185' )
plt.xlabel('Timepoints')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()
plt.xlim(0,46)
plt.show()
```

```python
capomulin_average_scatterplot=capomulin_df.groupby('Mouse ID').mean()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
plt.scatter(capomulin_average_scatterplot['Weight (g)'], capomulin_average_scatterplot['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin: Weight vs Tumor Volume')
plt.grid()
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation_coefficient= st.pearsonr(capomulin_average_scatterplot['Weight (g)'],capomulin_average_scatterplot['Tumor Volume (mm3)'])
correlation_coefficient
x_val=capomulin_average_scatterplot['Weight (g)']
y_val=capomulin_average_scatterplot['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr)=st.linregress(x_val,y_val)
regress_values= x_val*slope + intercept
line_eq='y= '+str(round(slope,2)) + 'x + ' + str(round(intercept,2))
plt.scatter(x_val,y_val)
plt.plot(x_val,regress_values, 'r-')
plt.annotate(line_eq,(20,35),fontsize=15,color="red")
plt.xlabel('Weight')
plt.ylabel('Tumor Volume')
plt.title('Mouse Weight vs Tumor Volume (mm3)')
plt.grid()
plt.show()
```
